```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 12, 6
```

```python
#df=pd.read_csv(r'Crime_Data_2010_2017')
df = pd.read_csv (r'C:\Users\hassan\Downloads\Crime_Data_2010_2017.csv')

```

```python
df[['Victim Age']].mean()
```

```python
victim = df[['Victim Age', 'Victim Sex', 'Victim Descent']]
```

```python
victim.isnull().any()
```

```python
age_fact = np.arange(victim['Victim Age'].min(), victim['Victim Age'].max(), 5)
plt.title('Crime on the basis of age')
plt.hist(victim['Victim Age'].dropna(), bins=age_fact, alpha=1, edgecolor='black')


                                                 
```

```python
sns.distplot(victim['Victim Age'].dropna(),5)
```

```python
plt.title('Victims on the basis of sex')
sns.countplot(victim['Victim Sex'])
```

```python
area = df[['Area ID', 'Area Name', 'Reporting District']]
```

```python
area.isnull().any()
```

```python
plt.title('Number of Crimes in Different Area')
sns.barplot(x='Area ID', data=area)
```
